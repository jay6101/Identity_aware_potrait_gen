```python
source = '/content/drive/Shareddrives/AML_Project/data/train_X'
dest = '/content/drive/Shareddrives/AML_Project/data/small_train_X'
files = os.listdir(source)
no_of_files = 200

for file_name in random.sample(files, no_of_files):
    shutil.copy(os.path.join(source, file_name), dest)
```
