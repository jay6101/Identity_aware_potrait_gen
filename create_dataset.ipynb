```python
import os
import random
import shutil
```

```python
source = '/content/drive/Shareddrives/AML_Project/data/copy_data_X/data_X'
dest = '/content/drive/Shareddrives/AML_Project/data/train_X'
files = os.listdir(source)
no_of_files = 1000

for file_name in random.sample(files, no_of_files):
    shutil.move(os.path.join(source, file_name), dest)
```
